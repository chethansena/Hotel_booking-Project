```python
#Analysize the cleaned data 
```

```python
cancelled_perc = df['is_canceled'].value_counts(normalize = True)
cancelled_perc
```

```python
#here we can see 62.8% of the reservations were not cancelled 
#around 37.1 % of the reservations were cancelled 
#this cancellation rate is very high, in the following i will visulize this so it will provide better picture, since the cancellation is half of not cancelled orders 
```

```python
plt.figure(figsize=(5,4))
plt.title("REsevation Status Count")
plt.bar(['Not canceled', 'Canceled'],df['is_canceled'].value_counts(), edgecolor = 'red', width = 0.7)
plt.show()
```

```python
plt.figure(figsize = (12,6))
ax1 = sns.countplot(x = 'hotel', hue = "is_canceled", data = df, palette = 'Blues', edgecolor='red')
legend_labels,_= ax1. get_legend_handles_labels()
#ax1.legend(bbox_to_anchor(1,1))
#ax1.legend(bbox_to_anchor=(1,1))
ax1.legend(loc='upper right')
plt.title("Reservation status in different hotels", size =20)
plt.xlabel('hotel')
plt.ylabel('number of reservations')
plt.legend(['not canceled', 'canceled'])
plt.show()
```

```python
#from above graph we can see city hotels have higher rate of cancelations as compared to resort hotels
```

```python
# to find the percentage of cancelations
resort_hotel = df[df['hotel']=='Resort Hotel']
resort_hotel['is_canceled'].value_counts(normalize = True)
```

```python
#for city hotels
city_hotel = df[df['hotel']=='City Hotel']
city_hotel['is_canceled'].value_counts(normalize = True)
#in this 41% is the cancelation rate
```

```python
resort_hotel = resort_hotel.groupby('reservation_status_date')[['adr']].mean()
city_hotel = city_hotel.groupby('reservation_status_date')[['adr']].mean()
```

```python
plt.figure(figsize=(20,8))
plt.title('Average Daily Rate in City and Resort Hotel', fontsize = 30)
plt.plot(resort_hotel.index, resort_hotel['adr'], label = 'Resort Hotel', color='blue')
plt.plot(city_hotel.index, city_hotel['adr'], label = 'City Hotel', color = 'red')
plt.legend(fontsize =20)
plt.show()
```

```python
df['month'] = df['reservation_status_date'].dt.month
plt.figure(figsize = (16,8))
ax1 = sns.countplot(x = 'month', hue = 'is_canceled', data=df, palette = 'bright', edgecolor ='red')
legend_labels,_= ax1. get_legend_handles_labels()
ax1.legend(bbox_to_anchor=(1,1))
plt.title("Reservation Status per Month", size=20)
plt.xlabel('Month')
plt.ylabel('Reservations')
plt.legend(['not_canceled', 'canceled'])
plt.show()

#as we can see highest number of cancellations has happened in the month of january and lowest in october
# here cancelations were more than reservations : most likely cause is price during these months 
```

```python
plt.figure(figsize=(15,8))
plt.title('ADR per Month', fontsize = 20)
sns.barplot('month','adr', data=df[df['is_canceled']==1].groupby('month')[['adr']].sum().reset_index())
plt.legend(fontsize=20)
plt.show()

#this proves that higher the price higher the cancelation rate. hence the cancelation rate is based on price of reservation
```

```python
canceled_data = df[df['is_canceled']==1]
top_10_country= canceled_data['country'].value_counts()[:10]
plt.figure(figsize=(8,8))
plt.title('Top 10 countries with highest Canclations')
plt.pie(top_10_country, autopct='%.2f', labels = top_10_country.index)
plt.show()

#portugal has higheest number of cancelations, it accounts to 70% of total cancelations 
```

```python
canceled_df_adr = canceled_data.groupby('reservation_status_date')[[('adr')]].mean()
canceled_df_adr.reset_index(inplace=True)
canceled_df_adr.sort_values('reservation_status_date', inplace = True)


not_canceled_data = df[df['is_canceled']==0]
not_canceled_df_adr = not_canceled_data.groupby('reservation_status_date')[[('adr')]].mean()
not_canceled_df_adr.reset_index(inplace=True)
not_canceled_df_adr.sort_values('reservation_status_date', inplace = True)





plt.figure(figsize=(20,6))
plt.title("Average Daily Rate", fontsize = 20)
plt.plot(not_canceled_df_adr['reservation_status_date'], not_canceled_df_adr['adr'], label = 'not_cancelled')
plt.plot(canceled_df_adr['reservation_status_date'], canceled_df_adr['adr'], label = 'canceled')
plt.legend
plt.show()
```
